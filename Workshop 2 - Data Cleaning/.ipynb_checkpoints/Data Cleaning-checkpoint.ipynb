```python
### Should we drop all the NAs in Times Pregnant column? 
```

```python
### Can we predict the number of Times Pregnant from the other columns? 
### HINT: use the Gender column to predict some of the Times Pregnan values
```

```python
diabetes[['Gender', 'Times Pregnant']]
```

```python
### impute 0s instead of NaNs for Male in the dataset
diabetes.loc[diabetes['Gender'] == 'Male', 'Times Pregnant'] = 0
```

```python
### check!
diabetes['Times Pregnant'].isnull().sum()
```

```python
### drop all the NaNs for Female
diabetes = diabetes.dropna(subset = ['Times Pregnant'])
```

```python
### YAY!!!
diabetes['Times Pregnant'].isnull().sum()
```

```python
### check for DUPLICATES!
### HINT: check using the unique idetifier
diabetes.duplicated(subset = 'id')
```

```python
### count the number of duplicates
diabetes.duplicated(subset = 'id').sum()
```

```python
### drop the duplicates
diabetes = diabetes.drop_duplicates(subset = 'id')
```

```python
### CHECK
diabetes['Times Pregnant'].isnull().sum()
### YAY!
```

```python
### LASTLY, convert the types!

yes_no_dict = {'Yes': 1,
              'No': 0}
diabetes = diabetes.replace(yes_no_dict)
diabetes.head()
```

```python
### you can also use pd.get_dummies
pd.get_dummies(diabetes['Gender'], drop_first = True)
```

```python
### add it as a column
diabetes['Gender'] = pd.get_dummies(diabetes['Gender'], drop_first = True)
diabetes
```
